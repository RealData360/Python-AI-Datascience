```python
import numpy as np #import the numpy library to peform numerical operations with your data
import pandas as pd #import the pandas library to create your dataframes

def warn(*args,**kwargs):
    pass
import warnings
warnings.warn = warn
warnings.filterwarnings('ignore')
```

```python
url = "https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29" #URL that contains our GDP data.
import requests #import the requests library so that you can retrieve data from the url. 
from bs4 import BeautifulSoup #import the BeautifulSoup library to so that you parse data from the url. 
import xml.etree.ElementTree as ET
```

```python
#Extract the required GDP data from the given URL using Web Scraping. Extract tables from webpage using Pandas, and retain table number 3 as the required dataframe.
GDP_tables = pd.read_html(url)
GDP_df = GDP_tables[3]
print(GDP_df)
```

```python
#Replace the column headers with column numbers
GDP_df2 = GDP_df.set_axis([0,1,2,3,4,5,6,7], axis ="columns")
GDP_df2
```

```python
# Retain columns with index 0 and 2 (name of country and value of GDP quoted by IMF)

GDP_df3 = GDP_df2.filter(items =[0,2])
GDP_df3
```

```python
# Retain the Rows with index 1 to 10, indicating the top 10 economies of the world.
GDP_df4 = GDP_df3.iloc[1:11]
GDP_df4
```

```python
# Assign column names as "Country" and "GDP (Million USD)"
GDP_df5 = GDP_df4.set_axis(['Country','GDP (Million USD)'], axis ="columns")
GDP_df5
```

```python
# Change the data type of the 'GDP (Million USD)' column to integer. Use astype() method.
GDP_df6 = GDP_df5.astype({'GDP (Million USD)':'int64'})
GDP_df6
```

```python
# Convert the GDP value in Million USD to Billion USD
GDP_df6['GDP (Billion USD)'] = np.round(GDP_df6['GDP (Million USD)']/1000, 2)
GDP_df6
```

```python
# Rename the column header from 'GDP (Million USD)' to 'GDP (Billion USD)'

GDP_df6 = GDP_df6.drop(columns=['GDP (Million USD)'])
GDP_df6
```

```python
# Load the DataFrame to the CSV file named "Largest_economies.csv"

GDP_CSV = GDP_df6.to_csv("Largest_Economies.csv")
GDP_CSV
```
